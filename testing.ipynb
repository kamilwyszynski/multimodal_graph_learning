```python
import networkx as nx
from stellargraph import StellarGraph

def get_stellar_graph(data):
    g = nx.Graph()

    for edge, sim in zip(data.edge_index_img[1::2], data.edge_attr_img[1::2]):
        node1 = data.y_img[int(edge[0])][0]
        node2 = data.y_img[int(edge[1])][0]

        feature1 = data.x_img[int(edge[0])].numpy()
        feature2 = data.x_img[int(edge[1])].numpy()

        g.add_edge(node1, node2, weight=float(sim), label='image2image')

        g.nodes[node1]['label'] = 'image'
        g.nodes[node2]['label'] = 'image'

        g.nodes[node1]['feature'] = feature1
        g.nodes[node2]['feature'] = feature2


    for edge, sim in zip(data.edge_index_wrd[1::2], data.edge_attr_wrd[1::2]):
        node1 = data.y_wrd[int(edge[0])][0]
        node2 = data.y_wrd[int(edge[1])][0]

        g.add_edge(node1, node2, weight=float(sim), label='word2word')
        g.nodes[node1]['label'] = 'word'
        g.nodes[node2]['label'] = 'word'

    
    for edge in data.edge_index_i2w:
        node1 = data.y_img[int(edge[0])][0]
        node2 = data.y_wrd[int(edge[1])][0]

        g.add_edge(node1, node2, label='image2word')
    
    return g
```
